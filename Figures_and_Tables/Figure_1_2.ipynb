```python
import wave, os, glob
import noisereduce
from scipy.io import wavfile
import matplotlib.pyplot as plt
import librosa.display
import sys
import numpy
import librosa
numpy.set_printoptions(threshold=sys.maxsize)
from pydub import AudioSegment,silence
import math
from scipy.stats import variation 
import statistics
import pandas as pd
import numpy as np
```

## Figure 1

```python
folder_with_files = #'...../Preprocessed_audio_files/AWNS' # Change path as required
```

```python
os.chdir(folder_with_files)
```

```python
#Figure 1a and 1b
wav_files = 'C5_HL.wav'
```

```python
startSec = [0.000] 
endSec   = [10]
startTimes = [] 
for s in startSec:
    startTimes.append(s*1000)     
endTimes = []
for e in endSec:
    endTimes.append(e*1000)    
for f, s, e in zip(wav_files, startTimes, endTimes):
    song = AudioSegment.from_wav(wav_files)
    s = int(s)           
    e = int(e)               
    extract = song[s:e]
    f = 'wav_file'
    extract.export(f+'_extract.wav', format="wav")  
```

```python
signal, fs = librosa.load("wav_file_extract.wav", sr= None) 
signal_n = signal / (2**15)
signal_len = len(signal_n)
FRAME_SIZE = 1028
HOP_LENGTH = 512
zcrs = librosa.feature.zero_crossing_rate(signal,frame_length=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
rms = librosa.feature.rms(y=signal,frame_length=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
frames = range(len(zcrs))
t = librosa.frames_to_time(frames,sr=fs, hop_length=512)
```

## 1a

```python
%matplotlib notebook
plt.figure(figsize=(10, 6))
plt.plot(t,rms, color="r",label='Root Mean Square Energy')

librosa.display.waveshow(signal, sr = fs)
plt.xlabel('Time (sec)',fontsize=15)
plt.ylabel('Amplitude',fontsize=15)
plt.legend(loc='best',prop={'size': 15})
```

## 1b

```python
plt.figure(figsize=(10, 6))
plt.plot(t,zcrs, color="r",label='Zero Crossing Rate')

librosa.display.waveshow(signal, sr = fs)
plt.xlabel('Time (sec)',fontsize=15)
plt.ylabel('Amplitude',fontsize=15)
plt.legend(loc='best',prop={'size': 15})
```
